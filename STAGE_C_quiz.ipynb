```python
# Convert 'TotalCharges' to numeric values
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'], errors='coerce')

# Fill missing values with 0
data['TotalCharges'].fillna(0, inplace=True)
```

```python
data['TotalCharges'] .dtype
```

```python

# Convert 'Churn' to binary values
data['Churn'] = data['Churn'].map({'No': 0, 'Yes': 1})

# Check the result
print(data['Churn'].head())
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import accuracy_score
```

```python
# List of categorical and numerical features
categorical = ['gender', 'SeniorCitizen', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'InternetService',
               'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies',
               'Contract', 'PaperlessBilling', 'PaymentMethod']
numerical = ['tenure', 'MonthlyCharges', 'TotalCharges']
```

```python
X = data[categorical + numerical]
y = data['Churn']

# Split the data into training and testing sets with an 80-20 split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
X_train.shape
```

```python
X.shape
```

```python
X_test.shape
```

```python
# Scale the numerical features
scaler = StandardScaler()
X_train_num = pd.DataFrame(scaler.fit_transform(X_train[numerical]), columns=numerical, index=X_train.index)
X_test_num = pd.DataFrame(scaler.transform(X_test[numerical]), columns=numerical, index=X_test.index)
```

```python
# One-hot encode the categorical features
encoder = OneHotEncoder(sparse_output=False, drop='first')  # drop='first' to avoid multicollinearity
X_train_cat = pd.DataFrame(encoder.fit_transform(X_train[categorical]), columns=encoder.get_feature_names_out(categorical), index=X_train.index)
X_test_cat = pd.DataFrame(encoder.transform(X_test[categorical]), columns=encoder.get_feature_names_out(categorical), index=X_test.index)
```

```python
# Combine scaled numerical and one-hot encoded categorical features
X_train_processed = pd.concat([X_train_num, X_train_cat], axis=1)
X_test_processed = pd.concat([X_test_num, X_test_cat], axis=1)
```

```python
# Define the models
models = {
    "RandomForest": RandomForestClassifier(random_state=1),
    "ExtraTrees": ExtraTreesClassifier(random_state=1),
    "XGBoost": XGBClassifier(random_state=1, use_label_encoder=False, eval_metric='logloss'),
    "LightGBM": LGBMClassifier(random_state=1)
}
```

```python
# Train and evaluate the models
results = {}
for name, model in models.items():
    model.fit(X_train_processed, y_train)
    y_pred = model.predict(X_test_processed)
    accuracy = accuracy_score(y_test, y_pred)
    results[name] = accuracy
    print(f"{name} Accuracy: {accuracy:.4f}")

# Display the results
print("Model evaluation results:")
for model_name, accuracy in results.items():
    print(f"{model_name}: {accuracy:.4f}")
```

```python
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.model_selection import RandomizedSearchCV

# Define ExtraTreesClassifier estimator
extra_trees = ExtraTreesClassifier(random_state=1)

# Define hyperparameter grid
n_estimators = [50, 100, 300, 500, 1000]
min_samples_split = [2, 3, 5, 7, 9]
min_samples_leaf = [1, 2, 4, 6, 8]
max_features = ['auto', 'sqrt', 'log2', None]
hyperparameter_grid = {
    'n_estimators': n_estimators,
    'min_samples_leaf': min_samples_leaf,
    'min_samples_split': min_samples_split,
    'max_features': max_features
}

# Perform RandomizedSearchCV
random_search = RandomizedSearchCV(extra_trees, hyperparameter_grid, cv=5, n_iter=10, scoring='accuracy', n_jobs=-1, verbose=1, random_state=1)
random_search.fit(X_train_processed, y_train)

# Extract and print the best hyperparameters
best_params = random_search.best_params_
print("Best Hyperparameters:")
print(best_params)
```

```python
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.metrics import accuracy_score

# Train the initial ExtraTreesClassifier model without hyperparameter tuning
initial_extra_trees = ExtraTreesClassifier(random_state=1)
initial_extra_trees.fit(X_train_processed, y_train)
initial_y_pred = initial_extra_trees.predict(X_test_processed)
initial_accuracy = accuracy_score(y_test, initial_y_pred)

# Extract the best hyperparameters from RandomizedSearchCV results
best_hyperparameters = random_search.best_params_

# Train a new ExtraTreesClassifier model with the best hyperparameters
optimal_extra_trees = ExtraTreesClassifier(random_state=1, **best_hyperparameters)
optimal_extra_trees.fit(X_train_processed, y_train)
optimal_y_pred = optimal_extra_trees.predict(X_test_processed)
optimal_accuracy = accuracy_score(y_test, optimal_y_pred)

# Compare the accuracies
print(f"Initial ExtraTreesClassifier Accuracy: {initial_accuracy:.4f}")
print(f"Optimal ExtraTreesClassifier Accuracy: {optimal_accuracy:.4f}")

if optimal_accuracy > initial_accuracy:
    print("The accuracy of the new optimal model is higher than the initial model.")
elif optimal_accuracy < initial_accuracy:
    print("The accuracy of the new optimal model is lower than the initial model.")
else:
    print("The accuracy of the new optimal model is the same as the initial model.")
```
